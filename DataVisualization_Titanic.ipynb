Task 3: Data Visualization on Titanic Dataset

Objective:
To create visual representations of Titanic passenger data and identify meaningful insights using charts and graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title('Passenger Survival Distribution')
plt.xlabel('0 = Not Survived, 1 = Survived')
plt.ylabel('Count')
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='bar')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
df['Age'].dropna().hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
pd.crosstab(df['Pclass'], df['Survived']).plot(kind='bar')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

Conclusion

1. Most passengers traveled in 3rd class.
2. Male passengers were more than female passengers.
3. Survival rate was lower than the death rate.
4. Female passengers had a higher chance of survival.
5. Passenger class influenced survival chances.
6. Data visualization helped identify important patterns and trends in the dataset.
